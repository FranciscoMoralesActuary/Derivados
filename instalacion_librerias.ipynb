```python
#Crear una funciÃ³n para la extracciÃ³n de datos de Yahoo Finance
def get_data(symbol=str):
    data=yf.download(tickers=symbol, period='252d', interval='1d')
    data.reset_index(inplace=True)
    return data
```

```python
df=get_data('VOLARA.MX')
df
```

```python
#Redefinir los encabezados
df=df.set_axis(['Date','Close','High','Low','Open','Volume'],axis=1)
df
```
