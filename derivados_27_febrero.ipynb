```python
#Importar Libreria de Yahoo Finance
import yfinance as yf
#Importar Libreria de Pandas
import pandas as pd
#Importar la Libreria para graficos
import plotly.graph_objects as go
```

```python
#Definir el Ticker de la Acción de la cual voy a descargar información, conforme a la estructura de Yahoo Finance
Ticker_YF='VOLARA.MX' #Acciones que revisarémos del mercado local: VOLARA.MX, CEMEXCPO.MX
#Crear el objeto Ticker
Ticker=yf.Ticker(Ticker_YF)
```

```python
#Revisar la información general de la empresa
info=Ticker.info
#print(info) # Una forma de imprimir la información del objeto
info
```

```python
#1ra Opción: Descarga de información dando un rango de fechas
Datos_Históricos_Op1=Ticker.history(start="2024-01-02",end="2025-02-27",interval="1d")
Datos_Históricos_Op1
```

```python
#2da Opción: Descargar información dando el número de días que quiero obtener de información
Datos_Historicos_Opc2=Ticker.history(period="252d",interval="1d")
Datos_Historicos_Opc2
```

```python
#Reindexar la tabla de datos
Datos_Historicos_Opc2.reset_index(inplace=True)
Datos_Historicos_Opc2
```

```python
#Eliminar el Time Zone y solo dejar la fecha
Datos_Historicos_Opc2['Date']=pd.to_datetime(Datos_Historicos_Opc2['Date']).dt.date
Datos_Historicos_Opc2
```

```python
datos=Datos_Historicos_Opc2[:]
grafico=go.Figure(data=go.Candlestick(x=datos['Date'],
                    open=datos['Open'],
                    high=datos['High'],
                    low=datos['Low'],
                    close=datos['Close'],
                    increasing_line_color='green',decreasing_line_color='red'))

grafico.show()
```

```python
#3ra opción: Extracción de información de datos
data=yf.download(tickers='VOLARA.MX', period='252d', interval='1d')
data
```

```python
#Crear una función para la extracción de datos de Yahoo Finance
def get_data(symbol=str):
    data=yf.download(tickers=symbol, period='252d', interval='1d')
    data.reset_index(inplace=True)
    return data
```

```python
df=get_data('VOLARA.MX')
df
```

```python
#Redefinir los encabezados
df=df.set_axis(['Date','Close','High','Low','Open','Volume'],axis=1)
df
```
