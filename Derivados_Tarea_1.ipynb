# Descargas de ETF's

Los **ETF (Exchange Traded Funds)** o Fondos Cotizados en Bolsa son fondos de inversión cuyo objetivo es replicar la evolución de un índice o de una cesta de valores.

Para descargarlos ocupamos la libreria **yfinance** que lo descarga directamente de Yahoo Finance solo con poner la abreviatura de la acción, para realizarlo se siguen los siguientes pasos





1.   **Importar yfinance**
2.   Definir el **Ticker** es la abreviatura con la que la acción aparace en Yahoo Finance
1.   Descargar la acciones con el comando **Ticker.history(start=$x$,end=$y$,interval=$z$)** donde $x$ es la fecha de inicio , $y$ fecha final (Esta día ya no se descarga) y $z$ es el intervalo (Minuto, Día , Semana y Mes )

2.   Indexamos la tabla con **reset_index**
1.   Convertimos el dato tipo "Time Zone" a "Date" de la columna $x$ con el comanndo **to_datetime(Datos_Históricos_Op1[$x$]).dt.date**
2.   Exportamos el dataframe en un excel con el comando **df1.to_excel('a.xlsx',index=False)** (Podemos nombrar el excel de la manera que desemos) y lo descargamos en la sección de archivos






# ETF de "Energy"

**First Trust NASDAQ Clean Edge Green Energy Index Fund ETF (QCLN)**: Invierte en empresas de energía limpia y renovable

El índice está diseñado para rastrear el rendimiento de las empresas de energía limpia de pequeña, mediana y gran capitalización que cotizan en bolsa en los Estados Unidos

image.png

```python
#Importar Libreria de Yahoo Finance
import yfinance as yf
#Importar Libreria de Pandas
import pandas as pd
#Importar la Libreria para graficos
import plotly.graph_objects as go
```

```python
#Definir el Ticker de la Acción de la cual voy a descargar información, conforme a la estructura de Yahoo Finance
Ticker_YF='QCLN'
#Crear el objeto Ticker
Ticker=yf.Ticker(Ticker_YF)
```

```python
#1ra Opción: Descarga de información dando un rango de fechas (Fecha de inicio, cierre e intervalor)
df_energy=Ticker.history(start="2021-12-31",end="2025-03-01",interval="1d")
df_energy
```

```python
#Reindexar la tabla de datos (inplace=True para que se aplica permanentemente en el dataframe)
df_energy.reset_index(inplace=True)
df_energy
```

```python
#Eliminar el Time Zone y solo dejar la fecha
df_energy['Date']=pd.to_datetime(df_energy['Date']).dt.date
df_energy
```

```python
#Seleccionamos la columnas 'Date' y 'Close'
df_energy=df_energy[['Date','Close']]
df_energy
```

```python
#Exportar los datos de la Acción a un excel
df_energy.to_excel('QCLN.xlsx',index=False)
```
